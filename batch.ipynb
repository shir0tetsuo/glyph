```python
from IPython.display import Image, display
import threading
lock = threading.Lock()
```

```python
import random
```

```python
from Components.loaders import SavedColors, SavedGlyphs
random.choice(list(SavedColors().maps.keys())) 
```

```python
def new_random(x:int=5, seed=42):
    set_of_new_uuids = [uuid.uuid4() for _ in range(x)]
    out_path = '/home/inpw/Pictures/gt/'
    for u in set_of_new_uuids:
        path_out = os.path.join(out_path, f'{u}.png')
        glyphtable = random.choice(list(SavedGlyphs().maps.keys()))
        cmap = random.choice(list(SavedColors().maps.keys()))
        with lock:
            generate_glyph_png(
                glyphtable=glyphtable,
                cmap=cmap,
                seed=seed,
                rows=1,
                cols=9,
                out_path=path_out,
                passed_uuid=u,
                #shorten_uuid=5,
                fsize=26
            )
        print(glyphtable, cmap, seed, u)
        
        display(Image(filename=path_out))
        
def new_set(glyphtable, cmap, x:int=5, seed=42):
    set_of_new_uuids = [uuid.uuid4() for _ in range(x)]
    out_path = '/home/inpw/Pictures/gt/'
    for u in set_of_new_uuids:
        path_out = os.path.join(out_path, f'{u}.png')
        with lock:
            generate_glyph_png(
                glyphtable=glyphtable,
                cmap=cmap,
                seed=seed,
                rows=1,
                cols=9,
                out_path=path_out,
                passed_uuid=u,
                #shorten_uuid=5,
                fsize=26
            )
            
        display(Image(filename=path_out))

```
